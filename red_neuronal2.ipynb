```python
import tensorflow as tf
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
```

```python
training_data = np.array([[11,0],[11,2],[11,1],[11,3],[11,4],[11,11],[11,0],[11,0],[11,2],[11,1],[11,3],[11,4],[11,11],[11,0]],'float32')
target_data = np.array([[0],[0.18],[0.09],[0.27],[0.36],[1],[0],[0],[0.18],[0.09],[0.27],[0.36],[1],[0]],'float32')

n_entradas = len(training_data[0])
n_nodos = 32
n_salida = 1
```

```python
model = Sequential()
model.add(Dense(n_nodos, input_dim = n_entradas, activation='relu'))
model.add(Dense(32, activation='sigmoid'))
model.add(Dense(32, activation='sigmoid'))
model.add(Dense(n_salida, activation='sigmoid'))

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['binary_accuracy'])
model.fit(training_data, target_data, epochs=200, verbose=1)
```

```python
score = model.evaluate(training_data, training_data, verbose=0)
print(f'  ... valor del modelo: {score[1]*100}%')
model.summary()
```

```python
real_data = np.array([[11,0]],'float32')
resultado = model.predict(real_data, verbose=1)
print(f'Entrada: {real_data[0]} => {resultado[0]}')
```
